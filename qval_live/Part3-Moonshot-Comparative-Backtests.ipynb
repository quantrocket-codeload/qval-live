# Moonshot Comparative Backtests

Next we will run a recent backtest using IB price data, and compare it to a comparable backtest using Sharadar price data.

This backtest serves as a sanity check: we researched and backtested with Sharadar price data but ultimately we will be trading using IB data, so we want to make sure the results are comparable. However, we limit the comparison to recent history: since Sharadar data includes delisted tickers and IB data does not, we naturally expect the backtest results to diverge more, the further back we go. For recent history this divergence should hopefully be small.

## qval-live strategy file

A variant of QVAL for live trading is implemented in [qval_live.py](qval_live.py). ## Run comparative backtests

We'll run the `qval` and `qval-live` backtests side-by-side to make it easy to compare them:

```python
from quantrocket.moonshot import backtest
%time backtest(["qval","qval-live"], start_date="2016-06-01", end_date="2018-01-01", segment="A", filepath_or_buffer="qval_comparative_backtest.csv")
```
